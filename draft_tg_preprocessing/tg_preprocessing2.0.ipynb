```python
column_names = ["time", "date", "message"]
data = pd.read_csv("filtered_translated.csv", names=column_names)
data
```

```python
def remove_emojis(text):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    text = re.sub(r'→', '', text)
    text = text.replace('"', '')
    return re.sub(emoj, '', text)
```

```python
def remove_links(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub('', text)
```

```python
data['message'] = data['message'].astype(str).apply(remove_emojis)
data
```

```python
data['message'] = data['message'].astype(str).apply(remove_links)
data
```

```python
data['message'] = data['message'].apply(lambda x: x.lower())
```

```python
aggregated = data.groupby('date')['message'].apply(' '.join).reset_index()
aggregated.to_csv('dataset-tg.csv', index=False)
```

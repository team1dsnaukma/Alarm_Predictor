```python
import pandas as pd
import pytz
```

```python
data = pd.read_csv("messages.csv")
data.tail(20)
```

```python
data['timestamp'] = data['date'] + ' ' + data['time']
data = data.drop(columns=['time', 'date'])
data = data[['timestamp', 'message']]
data.tail(20)
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'], utc=True).dt.tz_convert('Europe/Kiev')
data.tail(20)
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp']).dt.tz_localize(None)
data.tail(20)
```

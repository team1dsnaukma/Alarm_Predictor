```python
import pandas as pd
data = pd.read_csv("messages.csv")
data.head(20)
```

```python
data['timestamp'] = data['date'] + ' ' + data['time']
data = data.drop(columns=['time', 'date'])
data = data[['timestamp', 'message']]
data.head(20)
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['timestamp'] += pd.Timedelta(hours=2)
data.head(20)
```

```python
data.to_csv("tg-messages.csv", index=False)
```
